 * The article is understandable for someone, who does not have knowledge about machine learning. All indicators are precisely described, along with all the necessity formulas. For example, a reader does not have to search what F1 or accuracy is as it was all explained in the article.
 * The article was published on 14th May 2020. It was shortly after COVID-19 pandemic outbreak, when such models could be useful during the next months of pandemic. Also, the discussion connected with the article will be actual for a long time.

## Weak points of the article:
 * There is a huge difference between the accuracy mentioned in the article and in the research conducted in the review. In the article the overall accuracy was about 90%. The research on external data demonstrated in review “External validation demonstrates limited clinical utility of the interpretable mortality prediction model for patients with COVID-19 ”showed only 50% accuracy of the reproduced model.
 * Code in the repository is written partly in Chinese. This makes it difficult to understand, especially when variables are named in Chinese. It also causes the risk that there will be less reviews of the article and in consequence the research will be considered as less reliable.
